```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

connection_string = dbutils.secrets.get(scope="key-vault-scope", key="event-hub-conn")

eh_conf = {
    "eventhubs.connectionString": connection_string,
    # start fresh (latest events)
    "eventhubs.startingPosition": """{"offset": "@latest"}"""
}
```

```python
raw_df = (
    spark.readStream
         .format("eventhubs")
         .options(**eh_conf)
         .load()
)
```
